# Aggregate Data
- combine with temperature
- aggregate into zone 21 data

```python
import pandas as pd
data = pd.read_csv('../Data/Load_history.csv')
weather = pd.read_csv('../Data/temperature_history.csv')
data["date"] = pd.to_datetime(data[['year', 'month', 'day']])
weather['date'] = pd.to_datetime(weather[['year', 'month', 'day']])
total = []
column = [f'h{i}' for i in range(1, 25)]
stations = [weather[weather['station_id']==i+1] for i in range(11)]
```

```python
import numpy as np
total = []
data = data.values
for row in data:
    zone_id = row[0]
    date = row[-1]
    stat  = []
    for sid, station in enumerate(stations):
        r = station[station['date']==date].values
        if (len(r)==0):
            v = [np.nan for i in range(24)]
        else:
            v =  [r[0][4+i] for i in range(24)]
        stat.append(v)
    for i in range(24):
        delta = i
        t = date + pd.to_timedelta(i, unit='h')
        basic = [t, zone_id, row[4+i]]
        basic += [s[i] for s in stat]
        total.append(basic)
```

```python
df = pd.DataFrame(total)
df.columns = ['time', 'zone_id', 'value', 't1', 't2', 't3', 't4', 't5', 't6', 't7', 't8', 't9', 't10', 't11']
```
