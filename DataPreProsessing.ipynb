<h1> Data preprocessing

```python
import pandas as pd
import numpy as np
import os
import json
```

<h1> 1° Import data and parse as float and date

```python
df = pd.read_csv('./data/Load_history.csv', parse_dates={'datetime': ['year', 'month', 'day']}, index_col ='datetime')

#df = pd.read_csv('./data/Load_history.csv', index_col =['year', 'month', 'day'])
for col in df.columns[1:]:
    df[col] = df[col].apply(lambda x : str(x).replace(',','.'))
    df[col] = df[col].astype(float)
#df.sort_values(by = ['datetime', 'zone_id'])
```

```python
if not os.path.exists('./data/processed'):
    os.makedirs('./data/processed')
df.to_csv('./data/processed/full_data.csv')
df.head()
```

<h1> 2° process data by zone

```python

grouped_zone = df.groupby('zone_id')

if not os.path.exists('./data/processed/zoned'):
    os.makedirs('./data/processed/zoned')
for name, group in grouped_zone:
    group.to_csv('./data/processed/zoned/zone'+str(name)+'.csv')
print("files exported to ./data/processed/zoned")
```
